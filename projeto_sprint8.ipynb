### PRODUTO
**Iremos procurar responder as seguintes perguntas:**<br>
Quantas pessoas usam-no cada dia, semana e mês?<br>
Quantas sessões ocorrem por dia? (um usuário pode realizar várias sessões).<br>
Que comprimento tem cada sessão?<br>
Com que frequência os usuários voltam?<br>


````python
# Cálculando as métricas DAU (dia), WAU(semana) e MAU (mês)
visits_df["date"] = visits_df["start_ts"].dt.date
visits_df["week"] = visits_df["start_ts"].dt.to_period("W")
visits_df["month"] = visits_df["start_ts"].dt.to_period("M")

DAU = visits_df.groupby("date")["uid"].nunique()
WAU = visits_df.groupby("week")["uid"].nunique()
MAU = visits_df.groupby("month")["uid"].nunique()

# Número de sessões por dia
sessions_per_day = visits_df.groupby("date")["uid"].count()

# Duração média das sessões
visits_df["session_duration"] = (visits_df["end_ts"] - visits_df["start_ts"]).dt.total_seconds()
mean_session_duration = visits_df.groupby("date")["session_duration"].mean()

# Frequência de retorno dos usuários
returning_users = visits_df.groupby("uid")["date"].nunique().value_counts()

# Exibindo as métricas calculadas
print("DAU:")
print(DAU.describe())
print("\nWAU:")
print(WAU.describe())
print("\nMAU:")
print(MAU.describe())
print("\nNúmero de Sessões por Dia:")
print(sessions_per_day.describe())
print("\nDuração Média das Sessões:")
print(mean_session_duration.describe())
print("\nFrequência de Retorno dos Usuários:")
print(returning_users.head())

````

````python
# Exibindo os demais gráficos para uma melhor análise visual

plt.figure(figsize=(12, 5))
sns.lineplot(data=DAU, label="DAU")
sns.lineplot(data=WAU, label="WAU")
sns.lineplot(data=MAU, label="MAU")
plt.title("DAU, WAU e MAU ao longo do tempo")
plt.xlabel("Data")
plt.ylabel("Número de Usuários Únicos")
plt.legend()
plt.show()

plt.figure(figsize=(12, 5))
sns.lineplot(data=sessions_per_day)
plt.title("Número de Sessões por Dia")
plt.xlabel("Data")
plt.ylabel("Sessões")
plt.show()

plt.figure(figsize=(12, 5))
sns.lineplot(data=mean_session_duration)
plt.title("Duração Média das Sessões")
plt.xlabel("Data")
plt.ylabel("Segundos")
plt.show()
````

<strong>DAU(métrica diária)</strong> representa aproximadamente 4% de MAU (métrica mensal), o que indica que a maioria dos usuários não acessa o serviço diariamente.<br>
Já <strong>WAU(métrica semanal)</strong> é cerca de 25% de MAU, sugerindo que grande parte dos usuários acessa pelo menos uma vez por semana.<br>
E <strong>MAU(métrica mensal)</strong> permanece estável e crescente,indicando um fluxo constante de novos usuários, mas o DAU relativamente baixo pode indicar baixa retenção diária.<br>
Em relação ao <strong>número de sessões por dia</strong>, é muito próximo de 1, indicando que a maioria dos usuários não retorna várias vezes ao longo do dia.<br>
O valor máximo de 4042 sessões em um único dia sugere picos sazonais ou impacto de campanhas de marketing.<br>
Em relação a duração média das sessões, (cerca de 10 minutos) sugere que os usuários passam tempo considerável no serviço, o que pode indicar um bom nível de engajamento.<br>
Parece que há diferentes padrões de uso, ou seja, alguns usuários podem acessar rapidamente, enquanto outros ficam mais tempo navegando. <br>
Já em relação a <strong>Frequência de retorno dos usuários</strong>, a grande maioria dos usuários acessa apenas uma vez e nunca mais volta (<strong>ALERTA</strong>)!

Aqui você precisa calcular a taxa de retenção dos usuários por cohort (análise: "com que frequência os usuários voltam?"). A rotina é a seguinte:
    
1. Você precisa pegar a data do primeiro acesso de cada usuário
2. Para calcular o intervalo ou cohort de acesso para cada usuário, você precisa calcular o intervalo entre o acesso atual e o primeiro acesso:
    
```python
logs_first_log_month["cohort_age"] = logs_first_log_month["log_month"].astype(
    int)- logs_first_log_month["first_log_month"].astype(int)
```
 
3. A partir desses dados, você pode criar uma tabela dinâmica (`pivot_table`) que indexa por cada mês do conjunto de dados e calcula a quantidade de usuários que teve o primeiro acesso naquele mês e voltou a acessar no futuro (`cohort`): 1 mês depois, 2 meses depois, etc. 
4. A partir dessa tabela, você pode criar um gráfico de calor (`heatmap`), conforme sugestão abaixo.


</s>
</div>

````python
# Cálculo da Taxa de Retenção dos Usuários por Cohort
visits_df["first_log_month"] = visits_df.groupby("uid")["start_ts"].transform("min").dt.to_period("M")
visits_df["log_month"] = visits_df["start_ts"].dt.to_period("M")
visits_df["cohort_age"] = (visits_df["log_month"].astype(int) - visits_df["first_log_month"].astype(int))

cohort_counts = visits_df.pivot_table(index="first_log_month", columns="cohort_age", values="uid", aggfunc="nunique")
cohort_retention = cohort_counts.div(cohort_counts.iloc[:, 0], axis=0) * 100

#Criando um gráfico de calor
plt.figure(figsize=(12, 6))
sns.heatmap(cohort_retention, annot=True, fmt=".1f", cmap="coolwarm")
plt.title("Taxa de Retenção por Cohort (%)")
plt.xlabel("Meses Desde o Primeiro Acesso")
plt.ylabel("Mês da Coorte")
plt.show()
````

<div class="alert alert-info"><strong>As sugestões foram aplicadas.</strong> <br>
A retenção inicial é sempre 100% (mês 0), pois todos os usuários acessaram pelo menos uma vez.<br>
A retenção cai drasticamente após o primeiro mês, indicando que a maioria dos usuários não retorna após a primeira experiência.<br>
A retenção continua diminuindo ao longo do tempo, estabilizando em torno de 2% a 6% após alguns meses.<br>
Cohorts mais recentes (2018) mostram retenções ainda menores comparados a cohorts mais antigos.<br>

Parece que a maioria dos usuários acessa apenas uma vez e não retorna, com uma retenção muito baixa após 2 ou 3 meses.</div>

````python
# Criando a coluna 'month'
orders_df["month"] = orders_df["buy_ts"].dt.to_period("M")
costs_df["month"] = costs_df["dt"].dt.to_period("M")

# Cálculando o LTV, CAC e ROI
first_orders_df = orders_df.groupby("uid").agg({"buy_ts": "min"}).reset_index().rename(columns={"buy_ts": "first_purchase"})
first_orders_df["first_purchase_month"] = first_orders_df["first_purchase"].dt.to_period("M")

user_source = visits_df[["uid", "source_id"]].drop_duplicates()
first_orders_df = first_orders_df.merge(user_source, on="uid")

revenue_df = orders_df.groupby(["uid", "month"]).agg({"revenue": "sum"}).reset_index()
buyers = first_orders_df.merge(revenue_df, on="uid")

revenue_grouped_df = buyers.groupby(["source_id", "month"]).agg({"revenue": "sum"}).reset_index()
costs_grouped_df = costs_df.groupby(["source_id", "month"]).agg({"costs": "sum"}).reset_index()

report = pd.merge(costs_grouped_df, revenue_grouped_df, on=["source_id", "month"], how="left")
report["ltv"] = report["revenue"] / report["costs"]
report["cac"] = report["costs"] / report["revenue"]
report["roi"] = (report["ltv"] / report["cac"]) * 100


#Imprimindo o DataFrame resultante
print(report)

# Criando heatmap para visualização do LTV
ltv_cohort = report.pivot_table(index="month", columns="source_id", values="ltv", aggfunc="mean")
plt.figure(figsize=(12, 6))
sns.heatmap(ltv_cohort, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("LTV por Cohort e Origem de Tráfego")
plt.xlabel("Origem de Tráfego")
plt.ylabel("Mês da Coorte")
plt.show()
````

<div class="alert alert-info">As sugestões foram aplicadas. <br>
Podemos perceber pelo heatmap que a origem de tráfego com maior LTV foi em 1 em dezembro de 2017 (LTV = 8.04), o que sugere que essa campanha ou canal foi altamente eficaz.<br>
A maioria das células tem LTV abaixo de 5, sugerindo que a retenção e o ticket médio ainda são desafios.<br>
Origens de tráfego 9 e 10 apresentam um LTV consistentemente baixo, indicando que esses canais podem não estar trazendo clientes de alto valor.<br>

<strong>Problema identificado:</strong>
A grande variação do LTV entre os cohorts e origens de tráfego indica que nem todas as campanhas são igualmente eficazes.
Algumas campanhas podem estar trazendo muitos usuários de baixo valor, impactando o retorno sobre investimento (ROI).</div>

Muitas origens têm ROMI < 1, indicando perda de dinheiro.<br>
Origens 9 e 10 têm custo, mas não aparecem com bons ROMIs, sugerindo que não estão gerando clientes de alto valor.<br>
O maior orçamento foi gasto na origem 3, mas precisa ser validado se o retorno justificou esse custo alto.</div>

## CONSIDERAÇÕES FINAIS 

<div class="alert alert-info"><strong>Este projeto analisou o desempenho das estratégias de marketing através de métricas financeiras e de retenção. Utilizamos dados de visitas, compras e custos de anúncios para entender como os investimentos impactam as vendas e a fidelização dos clientes.</strong><br>

<strong>Os principais KPIs avaliados foram:</strong><br>
Lifetime Value (LTV): Receita média gerada por cliente ao longo do tempo.<br>
Custo de Aquisição de Cliente (CAC): Quanto custa trazer um novo cliente.<br>
Retorno sobre Investimento em Marketing (ROI e ROMI): Quanto retorno foi gerado a partir dos investimentos em anúncios.<br>
Retenção de Usuários: Com que frequência os clientes retornam após a primeira compra.<br>
<br>
<strong>Algumas campanhas foram altamente eficazes:</strong><br>
A origem 1 em dezembro de 2017 teve um ROMI de 64.58, o maior do estudo. Isso indica que os clientes adquiridos nesse período e canal geraram um ótimo retorno.<br>
Outras campanhas, como março de 2018 (origem 3 e 5), também tiveram bons resultados.
<br><br>
<strong>Muitos canais não foram rentáveis:</strong><br>
Diversos cohorts apresentaram ROMI abaixo de 1, indicando que o investimento nesses períodos não foi recuperado pelas vendas. <br>
As origens 9 e 10 tiveram ROMI consistentemente baixo, sugerindo que o tráfego vindo desses canais pode não ser qualificado. <br>
A origem 3 consumiu 43% do orçamento total, mas precisa ser validada em termos de retorno.<br>
O gasto com marketing aumentou gradativamente até dezembro de 2017 e depois começou a diminuir, indicando um possível ajuste nas campanhas ao longo do tempo.<br><br>
    
<strong>Retenção de clientes - desafio</strong><br>
Verificamos que a maioria dos clientes compra apenas uma vez e não retorna.<br>
O LTV médio é relativamente baixo, sugerindo que há espaço para aumentar o ticket médio ou incentivar compras repetidas.<br>
<br>
<strong> Recomendações para a equipe de Marketing</strong>
1. Redirecionar o orçamento para campanhas comprovadamente lucrativas.<br>
2. Melhorar a retenção de clientes e o ticket médio.<br>
3. Refinar a segmentação e estratégia de aquisição de clientes.<br>
4. Testar novas segmentações e públicos-alvo para identificar clientes com maior potencial de LTV.<br>
5. Analisar a qualidade do tráfego das origens 9 e 10 para entender se há um problema na conversão desses visitantes.<br>
6. Testar diferentes mensagens e ofertas para aumentar a conversão dos leads em compradores.<br>
